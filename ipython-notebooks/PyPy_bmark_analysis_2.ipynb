```python
loops_df["hit"] = map( lambda x: x.count if hasattr(x,"count") else 0, loops_df.loop )
loops_df["count"] = map( lambda x: len( x.operations ), loops_df.loop )

#foo = loops_df.loop[0]
#foo.count
```

```python
loops_df
```

```python
for i, loop in enumerate( loops_df.loop ):
    if i % 10 == 0:
        print "processing loop {}".format( i )
    exec_slice, guard_slice = strip_exec_guard( loop )
    # go through the loop again and add the counts
    num_guard     = 0
    num_exec      = 0
    num_use_guard = 0
    num_use_exec  = 0
    num_use_both  = 0
    num_unknown   = 0
    num_debug_mp  = 0
    num_guard_a     = 0
    num_exec_a      = 0
    num_use_guard_a = 0
    num_use_exec_a  = 0
    num_use_both_a  = 0
    num_unknown_a   = 0
    num_debug_mp_a  = 0
    asm_nlines = 0
    for op in loop.operations:
        if op.name == "debug_merge_point":
            num_debug_mp += 1
        elif op._is_guard:
            num_guard += 1
            num_guard_a += op._asm_nlines
        elif "use_both" in op.args:
            num_use_both += 1
            num_use_both_a += op._asm_nlines
        elif "use_g" in op.args:
            num_use_guard += 1
            num_use_guard_a += op._asm_nlines
        elif "unknown" in op.args:
            num_unknown += 1
            num_unknown_a += op._asm_nlines
        #elif "num_use_exec" in op.args:
        #    num_use_exec += 1
        else:
            num_exec += 1
            num_exec_a += op._asm_nlines
        try:
            asm_nlines += op._asm_nlines
        except AttributeError:
            pass
    # write the counts back to df
    loops_df.loc[i,"num_guard"] = num_guard
    loops_df.loc[i,"num_guard_a"] = num_guard_a
    loops_df.loc[i,"num_exec"]  = num_exec
    loops_df.loc[i,"num_exec_a"]  = num_exec_a
    loops_df.loc[i,"num_use_guard"] = num_use_guard
    loops_df.loc[i,"num_use_guard_a"] = num_use_guard_a
    loops_df.loc[i,"num_use_exec"] = num_use_exec
    loops_df.loc[i,"num_use_exec_a"] = num_use_exec_a
    loops_df.loc[i,"num_use_both"] = num_use_both
    loops_df.loc[i,"num_use_both_a"] = num_use_both_a
    loops_df.loc[i,"num_unknown"] = num_unknown
    loops_df.loc[i,"num_unknown_a"] = num_unknown_a
    loops_df.loc[i,"num_debug_mp"] = num_debug_mp
    loops_df.loc[i,"num_debug_mp_a"] = num_debug_mp_a
    loops_df.loc[i,"asm_nlines"] = asm_nlines
```

```python
cols = [ "count", "num_guard", "num_exec", "num_use_guard", "num_use_exec", "num_use_both", "num_unknown", "num_debug_mp",
    "asm_nlines", "num_guard_a","num_exec_a","num_use_guard_a","num_use_exec_a","num_use_both_a","num_unknown_a","num_debug_mp_a",]
slice_df = pd.DataFrame( columns=( ["arch", "bmark"] + cols ) )
for bmark in loops_df.bmark.unique():
    for arch in loops_df.arch.unique():

        _df = loops_df[ loops_df.bmark == bmark ][ loops_df.arch == arch ]
        weighted = {}
        for col in cols:
            weighted[col] = (_df["hit"] * _df[col]).sum()
            
        weighted["arch"] = arch
        weighted["bmark"] = bmark
        slice_df = slice_df.append( weighted, ignore_index=True )
            
            #if col.startswith("num_"):
            #    weighted[col] /= weighted["count"]
            
        #cols = ["arch", "bmark"] + cols
        
    
#weighted

slice_df
    
#(eparse_df["hit"] * eparse_df["num_guard"]).sum()
#(eparse_df["hit"] * eparse_df["num_use_guard"]).sum()
```
