```python
def get_asm_nlines( asm ):
    if asm is None or asm.isspace() or asm == "":
        return 0
    return asm.count( "\n" ) + 1

def op_type_hist( loops ):
    hist_dict = {}
    static_ops = 0
    dyn_ops = 0
    for l in loops:
        for o in l.operations:
            if o.name == "debug_merge_point":
                continue
            if o.name not in hist_dict:
                s, d = 0, 0
            else:
                s, d = hist_dict[ o.name ]
            hist_dict[ o.name ] = s + 1, d + l.count
            static_ops += 1
            dyn_ops += l.count
    return hist_dict, static_ops, dyn_ops

def asm_hist( loops, asm_dict={} ):
    static_ops = 0
    prev_asm_nlines = 0

    for l in loops:
        try:
            l.force_asm()
        except AttributeError:
            print "WARNING: no force_asm, skipping"
            
        # assign and normalize for the missing nlines
        zero_cnt = 0
        for i,o in reversed( list( enumerate( l.operations ) ) ):
            if o.name == "debug_merge_point":
              continue
            o.asm_nlines = get_asm_nlines( o.asm )
            if o.asm_nlines == 0:
                zero_cnt += 1
                #print "zero at idx", i
            elif zero_cnt:
                #print "nonzero at idx {} value {}".format( i, o.asm_nlines )
                new_asm_nlines = float( o.asm_nlines ) / (zero_cnt + 1)
                for j in xrange( i, i + zero_cnt + 1 ):
                    #print "unzeroing at idx {} to value {}".format( j, new_asm_nlines )
                    l.operations[j].asm_nlines = new_asm_nlines
                zero_cnt = 0

        for o in l.operations:
            if o.name == "debug_merge_point":
                continue
            prev_asm_nlines = get_asm_nlines( o.asm )
            # 
            s, a = asm_dict.get( o.name, (0, 0) )
            asm_dict[ o.name ] = s + 1, a + o.asm_nlines
            static_ops += 1
    return asm_dict, static_ops
```

```python
def plot_op_type_hist( hist_dict, bmark="?" ):
    df = pd.DataFrame( hist_dict.items(), columns=["op", "f"] )
    df['freq'], df['dyn_freq'] = zip( *df['f'] )
    del df['f']
    df = df.sort( columns="freq", ascending=False )
    #return df
    df.plot( kind='bar', x='op', y='freq', figsize=(14,6), title="static op frequency of " + bmark )
    df = df.sort( columns="dyn_freq", ascending=False )
    df.plot( kind='bar', x='op', y='dyn_freq', figsize=(14,6), title="dynamic op frequency of " + bmark )

    
def plot_asm_hist( asm_dict, bmark="?" ):
    df = pd.DataFrame( asm_dict.items(), columns=["op", "a"] )
    df["freq"], df["asm_cnt"] = zip( *df["a"] )
    df["avg_asm_cnt"] = df["asm_cnt"] / df["freq"]
    df = df.sort( columns="avg_asm_cnt", ascending=False )
    df.plot( kind='bar', x='op', y='avg_asm_cnt', figsize=(14,6), title="average asm/ir of " + bmark )
    
bmark = "ai"
hist_dict, _, _ = op_type_hist( all_loops["arm"][bmark][1] )
plot_op_type_hist( hist_dict, bmark=bmark )
asm_dict, _ = asm_hist( all_loops["arm"][bmark][1] )
plot_asm_hist( asm_dict, bmark=bmark )
```

```python
def get_mean_op_type_hist():
    bmarks = sorted( all_loops["arm"].keys() )
    hist = {}
    for bmark in bmarks:
        loops = all_loops["arm"][bmark][1]
        hist_dict, static_ops, dynamic_ops = op_type_hist( loops )
        for op, (freq, dyn_freq) in hist_dict.items():
            if op not in hist:
                s, d = 0.0, 0.0
            else:
                s, d = hist[ op ]
            hist[op] = s + float( freq ) / static_ops / len(bmarks), \
                        d + float( dyn_freq ) / dynamic_ops / len(bmarks)
    return hist
hist = get_mean_op_type_hist()
plot_op_type_hist( hist, bmark="all" )
```

```python
def categorize_hist( hist_dict ):
    cat = {"getfield_" : "memop",
           "setfield_" : "memop",
           "getarrayitem" : "memop",
           "setarrayitem" : "memop",
           "guard_"    : "guard",
           "int_"      : "int",
           "uint_"     : "int",
           "new"       : "new",
           "call"      : "call",
           "cond_call" : "call",
           "instance_ptr_eq" : "ptr",
           "ptr"       : "ptr",
           "unicode"   : "unicode",
           "str"       : "str",
           "copystr"   : "str",
           "copyunicode" : "unicode",
           "same_as"   : "int",
           "getinterior" : "memop",
           "setinterior" : "memop",
           "label"     : "ctrl",
           "jump"      : "ctrl",
           "finish"    : "ctrl",
           }
    new_hist_dict = {}
    for op, (sf, df) in hist_dict.items():
        for c_k, c_v in cat.items():
            if op.startswith( c_k ):
                op = c_v
                break
        if op in new_hist_dict:
            s, d = new_hist_dict[op]
        else:
            s, d = 0.0, 0.0
        new_hist_dict[op] = s + sf, d + df
    return new_hist_dict
plot_op_type_hist( categorize_hist( hist ), bmark="all" )
        
```
